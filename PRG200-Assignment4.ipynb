```python
#Simone Shree Pathak Baiju

inventory = {
    "rice": {"price": 120, "stock": 20},
    "milk": {"price": 90, "stock": 10},
    "bread": {"price": 60, "stock": 15},
    "eggs": {"price": 15, "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

def process_order(inventory, cart):
    total = 0
    print("---- Final Bill ----")
    for item, quantity in cart.items():
        if item not in inventory:
            print(f"Sorry, {item} not found in inventory")
        elif inventory[item]["stock"] < quantity:
            print(f"Sorry, not enough stock for {item}")
        else:
            cost = inventory[item]["price"] * quantity
            total += cost
            inventory[item]["stock"] -= quantity
            print(f"{item} x{quantity} = NPR {cost}")
    print(f"Grand Total: NPR {total}")
    print("--------------")
    print("Updated stock:", ", ".join(f"{k}={v['stock']}" for k, v in inventory.items()))

process_order(inventory, cart)
```

```python
#Simone Shree Pathak Baiju
def convert_date(date_str, from_cal, to_cal):
    year, month, day = date_str.split("-")
    year = int(year)
    if from_cal == to_cal:
        return date_str
    elif from_cal == "AD" and to_cal == "BS":
        year += 56
    elif from_cal == "BS" and to_cal == "AD":
        year -= 56
    return f"{year}-{month}-{day}"
    

def format_date(date_str, style, cal):
    bs_months = ["Baisakh","Jestha","Ashadh","Shrawan","Bhadra","Ashwin",
                 "Kartik","Mangsir","Poush","Magh","Falgun","Chaitra"]
    year, month, day = date_str.split("-")
    month_index = int(month) - 1
    month_name = bs_months[month_index] if cal == "BS" else month
    if style == "full":
        return f"{day}th {month_name}, {year} {cal}"
    elif style == "nepali":
        return f"{day} {month_name} {year} {cal}"
    else:
        return f"{date_str} {cal}"
        

customers = [
    {"name": "Ramesh Thapa", "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki", "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai", "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]
#Sahil Khan
for c in customers:
    converted = convert_date(c["date"], c["cal"], c["need"])
    formatted = format_date(converted, c["style"], c["need"])
    print(f"{c['name']} | Original: {c['date']} {c['cal']} | Converted: {formatted}")

```

```python
#Simone Shree Pathak baiju
def word_frequency(text):

    for ch in ".,!?":
        text = text.replace(ch, "")
    words = text.lower().split()
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) + 1
    
    top3 = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:3]
    print("Top 3 words are given below:")
    for word, count in top3:
        print(f"{word} — {count} times")

text = """ Nepal is a beautiful country. Nepal has Mount Everest.
Everest is the highest mountain in the world. Many tourists
visit Nepal every year to see Everest and other mountains.
Nepal is known for its mountains and natural beauty. """

word_frequency(text)
```

```python
#Simone Shree Pathak Baiju
accounts = {
    "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
    "A002": {"name": "Sunita Karki", "balance": 8500, "pin": "5678"},
    "A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
}

def atm(account_id, pin, action, amount=0):
    if account_id not in accounts:
        print("Account not found")
        return
    acc = accounts[account_id]
    if acc["pin"] != pin:
        print("Incorrect PIN")
        return
    if action == "balance":
        print(f"{acc['name']} | Balance: NPR {acc['balance']}")
    elif action == "deposit":
        acc["balance"] += amount
        print(f"Deposited NPR {amount}. New balance: NPR {acc['balance']}")
    elif action == "withdraw":
        if amount > acc["balance"]:
            print("Insufficient funds")
        else:
            acc["balance"] -= amount
            print(f"Withdrawn NPR {amount}. New balance: NPR {acc['balance']}")

atm("A001", "1234", "balance")
atm("A002", "0000", "withdraw", 2000)
atm("A002", "5678", "deposit", 3000)
atm("A003", "9012", "withdraw", 25000)
atm("A004", "1111", "balance")

```

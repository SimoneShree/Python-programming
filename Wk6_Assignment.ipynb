```python
#Simone Shree Pathak Baiju
import math

station_name = "Kathmandu Weather Station"
def get_average(temps):
    return sum(temps) / len(temps)

def get_deviation(temps):
    mean = get_average(temps)  # local variable
    variance = sum((t - mean) ** 2 for t in temps) / len(temps)
    return math.sqrt(variance)

def get_summary(temps):
    print(f"Station: {station_name}")
    print(f"Min: {min(temps)}")
    print(f"Max: {max(temps)}")
    print(f"Average: {get_average(temps):.2f}")
    print(f"Deviation: {get_deviation(temps):.2f}")
```

```python
#Simone Shree Pathak Baiju
temperatures = [18.4, 22.1, 15.7, 29.3, 11.8, 25.6, 19.2]
get_summary(temperatures)
```

```python
#Simone Shree Pathak Baiju
import random

random.seed(42)
def split_bill(friends, total):
    return total / len(friends)

def pick_lucky(friends):
    return random.choice(friends)

def final_summary(friends, total):
    share = split_bill(friends, total)
    lucky = pick_lucky(friends)
    lucky_total = share + 50

    print(f"Total Bill: NPR {total}")
    print(f"Each person's share: NPR {share:.2f}")
    print()
    for friend in friends:
        print(f"{friend}: NPR {share:.2f}")
    print(f"\nLucky tax: {lucky} pays an extra NPR 50")
    print(f"{lucky}'s total: NPR {lucky_total:.2f}")
```

```python
#Simone Shree Pathak Baiju
friends = ["Ramesh", "Sunita", "Bikash", "Anjali", "Dipak"]
total_bill = 3750
final_summary(friends, total_bill)
```

```python
#Simone Shree Pathak Baiju
import datetime

college_name = "Bhaktapur Multiple Campus"
def parse_date(date_str):
    return datetime.datetime.strptime(date_str, "%Y-%m-%d")

def get_exam_date(start_str, days):
    start = parse_date(start_str)
    exam = start + datetime.timedelta(days=days)
    return exam.strftime("%Y-%m-%d")

def print_schedule(start_str, exams):
    print(f"College: {college_name}")
    print(f"{'Subject':<25} {'Exam Date'}")
    print("-" * 40)
    for subject, days in exams:
        date = get_exam_date(start_str, days)
        print(f"{subject:<25} {date}")
```

```python
#Simone Shree Pathak Baiju
start_date = "2025-05-01"
exams = [
    ("Python Programming", 0),
    ("Data Structures", 3),
    ("Database Systems", 6),
    ("Computer Networks", 10),
    ("Mathematics", 14),
]

print_schedule(start_date, exams)
```
